## Number of Revenue-generating Sessions

Out of 12330 sessions, how many actually converted to generating revenue?


````python
# Count the number of sessions for each value in the 'Revenue' column
revenue_counts = data['Revenue'].value_counts()

# Plot the counts as a bar chart
revenue_counts.plot(kind='bar', color=['skyblue', 'orange'], edgecolor='black')
plt.title('Number of Revenue-Generating Sessions')
plt.xlabel('Revenue')
plt.ylabel('Number of Sessions')
plt.xticks(ticks=[0, 1], labels=['No Revenue', 'Revenue'], rotation=0)
plt.show()
````

````python
# Plot histograms for numerical columns
data.hist(figsize=(15, 10), bins=20, edgecolor='black')
plt.suptitle('Histograms of Numerical Columns')
plt.show()
````

````python
# Display value counts for categorical columns
print("\nValue Counts for Categorical Columns:")
for column in data.select_dtypes(include=['object']).columns:
    print(f"\n{column} Value Counts:")
    print(data[column].value_counts())
````

````python
# Pairplot for numerical columns
sns.pairplot(data.select_dtypes(include=['float64', 'int64']))
plt.suptitle('Pairplot of Numerical Columns', y=1.02)
plt.show()
````

Most obvious correlation graph is between 'BounceRates' and 'ExitRates', which makes sense.

People may have one glance at the website, and realised they weren't interested in it, and click away.


````python
# Out of curiousity, what's the correlation between 'BounceRates' and 'ExitRates'?

correlation = data['BounceRates'].corr(data['ExitRates'])
print(f"Correlation between BounceRates and ExitRates: {correlation}")
````

As expected, quite high
